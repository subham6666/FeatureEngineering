```python
me = pd.read_csv('Mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
me.head()
```

```python
## Lets see how many different categories we have 

me.X1.value_counts().sort_values(ascending=False)
```

```python
## Now if i want to apply one hot encoding for XI then according To KDD orange cup challenge(Kaggle)
## Use of Top 10 features

lst_10 = me.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
    me[categories]=np.where(me['X1']==categories,1,0)
```
